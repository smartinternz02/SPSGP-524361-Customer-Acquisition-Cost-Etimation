```python
#Perform Below Visualizations.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('titanic.csv')

# Univariate Analysis
# Example: Histogram of Age
plt.figure(figsize=(8, 6))
sns.histplot(data['age'].dropna(), kde=True)
plt.xlabel('age')
plt.ylabel('Count')
plt.title('Distribution of Age')
plt.show()

# Bi-Variate Analysis
# Example: Bar plot of Survival by Passenger Class
plt.figure(figsize=(8, 6))
sns.countplot(x='pclass', hue='survived', data=data)
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Survival by Passenger Class')
plt.show()

# Multi-Variate Analysis
# Example: Scatter plot of Age and Fare colored by Survival
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='fare', hue='survived', data=data)
plt.xlabel('age')
plt.ylabel('fare')
plt.title('Relationship between Age, Fare, and Survival')
plt.show()
```

```python
#Handle the Missing values.
import pandas as pd

# Load the dataset
data = pd.read_csv('titanic.csv')

# Check for missing values
print(data.isnull().sum())

# Option 1: Drop rows with missing values
data_dropped = data.dropna()
print("Dataset after dropping missing values:")
print(data_dropped.head())

# Option 2: Fill missing values with a specific value
data_filled = data.fillna(value=0)
print("Dataset after filling missing values:")
print(data_filled.head())

# Option 3: Fill missing values with mean, median, or mode
data_filled_mean = data.fillna(data.mean())
data_filled_median = data.fillna(data.median())
data_filled_mode = data.fillna(data.mode().iloc[0])
```

```python
#Find the outliers and replace the outliers
import pandas as pd
import numpy as np
from scipy import stats

# Load the dataset
data = pd.read_csv('titanic.csv')

# Identify outliers using z-score
z_scores = np.abs(stats.zscore(data['age']))
threshold = 3  # Set the threshold for determining outliers
outliers = np.where(z_scores > threshold)[0]

# Replace outliers with median value
median_age = data['age'].median()
data.loc[outliers, 'age'] = median_age

# Check if outliers are replaced
print("Outliers replaced:")
print(data.loc[outliers, 'age'])
```

```python
#Check for Categorical columns and perform encoding.
import pandas as pd

# Load the dataset
data = pd.read_csv('titanic.csv')

# Check for categorical columns
categorical_columns = data.select_dtypes(include=['object']).columns
print("Categorical columns:")
print(categorical_columns)

# Perform one-hot encoding using get_dummies()
encoded_data = pd.get_dummies(data, columns=categorical_columns)
print("Encoded data:")
print(encoded_data.head())
```

```python
#Split the data into dependent and independent variables.
import pandas as pd

# Load the dataset
data = pd.read_csv('titanic.csv')

# Split into dependent and independent variables
X = data.drop('survived', axis=1)  # Independent variables
y = data['survived']  # Dependent variable

# Print the shapes of X and y
print("Independent variables (X) shape:", X.shape)
print("Dependent variable (y) shape:", y.shape)
```

```python
from sklearn.preprocessing import StandardScaler

# Scale the independent variables
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)

# Print the scaled independent variables
print("Scaled independent variables:")
print(x_scaled)
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)

# Print the shapes of the training and testing sets
print("Training set - X shape:", x_train.shape)
print("Training set - y shape:", x_train.shape)
print("Testing set - X shape:", x_test.shape)
print("Testing set - y shape:", x_test.shape)
```
